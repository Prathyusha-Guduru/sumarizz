```python
import pandas as pd
```

```python
df = pd.read_csv('article.csv')
```

```python
import torch
from keybert import KeyBERT

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
kw_model = KeyBERT(model='all-MiniLM-L6-v2')
```

```python
df['keywords'] = df['article'].apply(
    lambda x: kw_model.extract_keywords(
        x,
        keyphrase_ngram_range=(1, 1),
        top_n=20
    )
)
```

```python
df.head()
```

```python
df.iloc[0]['keywords']
```

```python
## make a new column with just keywords
df['keywords_with_prob'] = df['keywords']
df['keywords'] = df['keywords'].apply(lambda x: [i[0] for i in x])
```

```python
# save df to csv
df.to_csv('article_with_keywords.csv', index=False)
```
